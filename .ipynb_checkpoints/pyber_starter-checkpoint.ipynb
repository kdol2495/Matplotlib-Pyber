```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
citydata_df = pd.read_csv(city_data_to_load)
citydata_df.head()

ridedata_df = pd.read_csv(ride_data_to_load)
ridedata_df.head()
# Combine the data into a single dataset
mergedata_df = pd.merge(citydata_df, ridedata_df)
# Display the data table for preview
mergedata_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
# Build the scatter plots for each city 


urbandata = mergedata_df.loc[mergedata_df["type"] == "Urban", :]
suburbandata = mergedata_df.loc[mergedata_df["type"] == "Suburban", :]
ruraldata = mergedata_df.loc[mergedata_df["type"] == "Rural"]

# grouping
groupurban = urbandata.groupby("city")
groupsuburban = suburbandata.groupby("city")
grouprural = ruraldata.groupby("city")

# urban
urbanfare = groupurban["fare"].mean()
urbanrides = groupurban["ride_id"].count()
urbancount = groupurban["driver_count"].sum()
urbandrivers = (urbancount // urbanrides)


# suburban
suburbanfare = groupsuburban["fare"].mean()
suburbanrides = groupsuburban["ride_id"].count()
suburbancount = groupsuburban["driver_count"].sum()
suburbandrivers = (suburbancount // suburbanrides)
# rural
ruralfare = grouprural["fare"].mean()
ruralrides = grouprural["ride_id"].count()
ruralcount = grouprural["driver_count"].sum()
ruraldrivers = (ruralcount // ruralrides)

# doing another dataframe thing
urbandata_df = pd.DataFrame({"AverageFare": urbanfare,
                       "TotalRides": urbanrides,
                       "TotalDrivers": urbandrivers})
urbandata_df["AverageFare"] = urbandata_df["AverageFare"].round(2)

suburbandata_df = pd.DataFrame({"AverageFare": suburbanfare,
                     "TotalRides": suburbanrides,
                     "TotalDrivers": suburbandrivers})
suburbandata_df["AverageFare"] = suburbandata_df["AverageFare"].round(2)

ruraldata_df = pd.DataFrame({"AverageFare": ruralfare,
                       "TotalRides": ruralrides,
                       "TotalDrivers": ruraldrivers})
ruraldata_df["AverageFare"] = ruraldata_df["AverageFare"].round(2)

# Incorporate the other graph properties
ChangeBubbleSize = 12
CBS = ChangeBubbleSize
plt.figure(figsize=(10,8))

# other bubbleplot formatting stuff

buburban = plt.scatter(urbandata_df.TotalRides, urbandata_df.AverageFare, c="LightCoral", label = "Urban", 
                    s=(urbandata_df.TotalDrivers * CBS), alpha = 0.7, edgecolor = "black", linewidths = 0.7)

# Suburban Scatter Plot Properties
bubsuburban = plt.scatter(suburbandata_df.TotalRides, suburbandata_df.AverageFare, c="LightSkyBlue", label = "Suburban", 
                    s=(suburbandata_df.TotalDrivers * CBS), alpha = 0.7, edgecolor = "black", linewidths = 0.7)

# Rural Scatter Plot Properties
bubrural = plt.scatter(ruraldata_df.TotalRides, ruraldata_df.AverageFare, c="gold", label = "Rural", 
                    s=(ruraldata_df.TotalDrivers * CBS), alpha = 0.7, edgecolor = "black", linewidths = 0.7)

lgnd = plt.legend(handles=[buburban, bubsuburban, bubrural], scatterpoints=1, fontsize=14)
lgnd.legendHandles[0]._sizes = [150]
lgnd.legendHandles[1]._sizes = [150]
lgnd.legendHandles[2]._sizes = [150]

# Create a legend
plt.title("Pyber Ride Sharing Data (2018)")
plt.xlabel("Total Number of Rides Per City")
plt.ylabel("Average Fare ($)")
# Incorporate a text label regarding circle size
plt.figtext(.95, .5, "Note: \nCircle size reflects number of drivers per city", fontsize=12)
# Save Figure
plt.savefig('pyber_bubbleplot.png')

plt.grid()
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
group_mergedata_df = mergedata_df.groupby('type')

faretotal = mergedata_df['fare'].sum()
group_fare = group_mergedata_df['fare'].sum()

group_faretotal = group_fare / faretotal
# Build Pie Chart
explode = (0, 0, 0.15)
labels = group_faretotal.index.tolist()
colors = ["gold", "LightSkyBlue", "LightCoral"]
plt.figure(figsize=(6,4))
plt.rcParams['font.size'] = 12.0
plt.pie(group_faretotal, labels = labels, explode = explode, colors = colors, startangle = 150, shadow = True, autopct="%1.1f%%")
plt.title("Percent of Total Fare by City Type")

# Save Figure
plt.savefig("totalfaresbycity.png")
plt.show()
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents

ridetotal = mergedata_df["ride_id"].count()
group_ride = group_mergedata_df["ride_id"].count()

group_ridetotal = group_ride / ridetotal

# Build Pie Chart
explode = (0, 0, 0.15)
labels = group_ridetotal.index.tolist()
colors = ["gold", "LightSkyBlue", "LightCoral"]
plt.figure(figsize=(6,4))
plt.rcParams['font.size'] = 12.0
plt.pie(group_ridetotal, labels = labels, explode = explode, colors = colors, startangle = 150, shadow = True, autopct="%1.1f%%")
plt.title("Percent of Total Rides by City Type")

# Save Figure
plt.savefig("totalridesbycity.png")
plt.show()
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
group_citydata_df = citydata_df.groupby("type")

drivertotal = citydata_df["driver_count"].sum()
group_driver = group_citydata_df["driver_count"].sum()

group_drivertotal = group_driver / drivertotal
# Build Pie Charts
explode = (0, 0, 0.15)
labels = group_drivertotal.index.tolist()
colors = ["gold", "LightSkyBlue", "LightCoral"] 
plt.figure(figsize=(6,4))
plt.rcParams['font.size'] = 12.0
plt.pie(group_drivertotal, labels = labels, explode = explode, colors = colors, startangle = 150, shadow = True, autopct="%1.1f%%")
plt.title("Percent of Total Drivers by City Type")

# Save Figure
plt.savefig("totaldriversbycitytype.png")
plt.show()
```

```python
# Observations and Insights about the data

# 1. Despite rural areas comprising far less of the Pyber market than suburban and urban areas, it's interesting that rural areas make up much larger proportions of the total fare and total rides than the proportion of total drivers. Perhaps there's more access in rural areas to Pyber riding services than there is access to Pyber employment opportunities as a driver. 

# 2. On the other hand urban areas make up a more significant proportion of drivers than the ubran proportion of rides and fares. Perhaps there is an inverse quality to urban access to Pyber driving employment than the hypthesized phenomenon in #1, where people who live in urban areas have far greater access to Pyber employment opportunities than suburban and rural areas. 

# 3. The range of values for average fare (the vertical range in the bubble plot) is far wider for rural areas than for suburban and urban areas. This could be due to the smaller sample size in rural areas, reflected in the smaller size of the rural bubbles and the fewer overall quantity of rural bubbles in comparison to the suburban and urban bubbles in the bubble plot.
```
